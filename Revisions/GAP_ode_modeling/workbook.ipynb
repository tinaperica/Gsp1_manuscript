Tanja:
The NMR data suggest that the two observable Gsp1 states g1 and g2 are in slow exchange.  The Geyer et al paper gives rather broad estimates of the exchange rate with an upper limit of approx 200 sec-1 and a lower limit of approx 10-3 sec-1.  If we assume a rate on the slower end, it would be much slower than the observed kcat (~5-10 sec-1).  So if hydrolysis is much faster than Gsp1 state interconversion, and if g2 is the catalytically relevant conformation of Gsp1, effectively only the g2 population of Gsp1 is available for the reaction.  With these assumptions, the Km - substrate concentration needed for 0.5 vmax - should be proportional to the fraction of total Gsp1 in the g2 state. Since we have estimates of the rates, perhaps it makes sense to model to see the effect of the assumptions, Chris, what do you think?

```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", context="talk")

import ode_modeler
from ode_modeler import io as io
from ode_modeler.runmodel import run


# set arguments
kwargs = {'species_file': 'species.csv',
          'ode_file': 'ode.csv',
          'queries_file': 'queries.csv',
          'out_dir': io.prepare_out('.'),
          'n_iter': 100,
          'ttot': 6000,
          'solver_method': 'LSODA' 
         }
```

```python
# import species information
species = io.import_species(**kwargs)

total_Gsp1 = (species['Gsp1GTP_II'].conc0 +
              species['Gsp1GTP_I'].conc0 +
              species['Gsp1GDP'].conc0)

# create a WT query
query = ode_modeler.io.Query()
query.name = 'WT'

# update parameters, for WT the steady state between states II and I
# should by p1/p2 = 87/13 = 6.7
query.parameters.update({'kcat':8.9, 'Km':0.4, 'k_ex_r':0.001/6.7, 'k_ex_f':0.001})


# define a function to run a single query many times with one
#   parameter systematically changed
def sweep_param(query, param, coef):
    queries = []
    
    for adj in coef:
        q = query.deep_copy()
        q.parameters[param] *= adj
        q.name = np.round(q.parameters[param], 2)
        q.df = run(species, q.parameters, **kwargs)
        queries.append(q)
    
    df = pd.concat(objs=[q.df for q in queries],
                   keys=[q.name for q in queries],
                   names=('param_value',))
    df.reset_index(level=0, inplace=True)
    df = df.assign(t = df.t/60,
                   Gsp1GDP = df.Gsp1GDP/total_Gsp1,
                   frac_II = df.Gsp1GTP_II/(df.Gsp1GTP_I+df.Gsp1GTP_II),
                   swept_param = param)

    return df


# set coefficients that sweep the parameters
kcat_sweep = np.logspace(-1, 1, num=9)
Km_sweep = np.logspace(-3, 5, num=9, base=2)

# prepare a list for dataframes
dfs = []

# with conformations in slow exchange, sweep kcat and Km
ex = 'slow exchange, k_ex_f = 0.001'
dfs.append(sweep_param(query, 'kcat', kcat_sweep).assign(exchange = ex))
dfs.append(sweep_param(query, 'Km', Km_sweep).assign(exchange = ex))

# update to medium exchange
query.parameters.update({'k_ex_f':0.01, 'k_ex_r':0.01/6.7})
ex = 'medium exchange, k_ex_f = 0.01'

dfs.append(sweep_param(query, 'kcat', kcat_sweep).assign(exchange = ex))
dfs.append(sweep_param(query, 'Km', Km_sweep).assign(exchange = ex))

# update to fast exchange
query.parameters.update({'k_ex_f':200, 'k_ex_r':200/6.7})
ex = 'fast exchange, k_ex_f = 200'

dfs.append(sweep_param(query, 'kcat', kcat_sweep).assign(exchange = ex))
dfs.append(sweep_param(query, 'Km', Km_sweep).assign(exchange = ex))

# combine all dataframes
df = pd.concat(dfs).reset_index(drop=True)
```

```python
### PLOTTING

# function to plot each set of parameter swept progress curves
def plot_fn(df, param, ex, ax1, ax2):
    
    data=df.loc[(df.swept_param == param) & (df.exchange == ex)]
    
    # set palettes for the two parameter sweeps
    # note that red is slower kcat, weaker Km
    if param == 'kcat':
        palette = sns.color_palette("coolwarm", 9)[::-1]
        palette[4] = [0,0,0]  # set WT to black

    if param == 'Km':
        palette = sns.color_palette("coolwarm", 11)[2:]
        palette[3] = [0,0,0]  # set WT to black
    
    # PROGRESS CURVE
    sns.lineplot(data=data, x = 't', y = 'Gsp1GDP', hue = 'param_value',
                 ax = ax1, palette=palette, linewidth=2, legend='full')
    
    ax1.legend(title=param, loc=2,
              handles=ax1.get_legend_handles_labels()[0][1:],
              labels=ax1.get_legend_handles_labels()[1][1:], # gets rid of the hue variable name
              bbox_to_anchor=(1.05, 1), borderaxespad=0.)
    
    ax1.set(xlabel='Time (min)', ylabel='Fraction Gsp1GTP hydrolyzed',
            title='GAP progress curve, '+ ex)
    
    
    # FRACTION IN STATE II
    sns.lineplot(data=data, x = 't', y = 'frac_II', hue = 'param_value',
                 ax = ax2, palette=palette, linewidth=2, legend='full')
    ax2.legend(title=param, loc=2,
              handles=ax2.get_legend_handles_labels()[0][1:],
              labels=ax2.get_legend_handles_labels()[1][1:], # gets rid of the hue variable name
              bbox_to_anchor=(1.05, 1), borderaxespad=0.)
    ax2.set(xlabel='Time (min)', ylabel='Fraction of remaining Gsp1GTP',
            xlim=(0,100), ylim=(0,1),
            title='Gsp1GTP in State II, '+ ex)

# plotting for kcat
fig, axs = plt.subplots(2, 3, figsize=(30, 15))
plot_fn(df, 'kcat', 'slow exchange, k_ex_f = 0.001', ax1 = axs[0,0], ax2 = axs[1,0])
plot_fn(df, 'kcat', 'medium exchange, k_ex_f = 0.01', ax1 = axs[0,1], ax2 = axs[1,1])
plot_fn(df, 'kcat', 'fast exchange, k_ex_f = 200', ax1 = axs[0,2], ax2 = axs[1,2])
fig.tight_layout()
plt.savefig('plots/kcat_sweep.png')

# plotting for Km
fig, axs = plt.subplots(2, 3, figsize=(30, 15))
plot_fn(df, 'Km', 'slow exchange, k_ex_f = 0.001', ax1 = axs[0,0], ax2 = axs[1,0])
plot_fn(df, 'Km', 'medium exchange, k_ex_f = 0.01', ax1 = axs[0,1], ax2 = axs[1,1])
plot_fn(df, 'Km', 'fast exchange, k_ex_f = 200', ax1 = axs[0,2], ax2 = axs[1,2])
fig.tight_layout()
plt.savefig('plots/Km_sweep.png')
```

# Now, simulate all curves, with slow and fast exchange

I tried with medium exchange so k_ex_f = 1 and 1e-1, those are the same as fast exchange

1e-2 starts to show differences

```python
# define a function to run each mutant with slow and fast exchange
def run_mutant(mutant, kcat, Km, NMR_ratio, NMR_frac_II):
    
    dfs = []

    # import species information
    species = io.import_species(**kwargs)

    
    # set the initial concentrations based on the NMR distributions
    total_Gsp1 = 8
    species['Gsp1GTP_II'].conc0 = total_Gsp1*NMR_frac_II
    species['Gsp1GTP_I'].conc0 = total_Gsp1*(1-NMR_frac_II)
    
    q = ode_modeler.io.Query()
    q.name = mutant
    q.parameters.update({'kcat':kcat, 'Km':Km})

    # update parameters for mutant, and start with slow exchange
    # exchange rates cannot be the same, otherwise all mutants would be 50:50 at eq.
    
    if not np.isfinite(NMR_ratio):  # if biased state 2
        q.parameters.update({'k_ex_f':1e-3, 'k_ex_r':1e-5})
    elif NMR_ratio == 0.0:  # if biased state 1
        q.parameters.update({'k_ex_f':1e-5, 'k_ex_r':1e-3})
    else: 
        q.parameters.update({'k_ex_f':1e-3, 'k_ex_r':1e-3/NMR_ratio})
        
    df = run(species, q.parameters, **kwargs)
    df = df.assign(exchange = 'slow exchange, k_ex_f = 0.001')

    dfs.append(df)

    # run with medium exchange
    if not np.isfinite(NMR_ratio):  # if biased state 2
        q.parameters.update({'k_ex_f':1e-2, 'k_ex_r':1e-4})
    elif NMR_ratio == 0.0:  # if biased state 1
        q.parameters.update({'k_ex_f':1e-4, 'k_ex_r':1e-2})
    else: 
        q.parameters.update({'k_ex_f':1e-2, 'k_ex_r':1e-2/NMR_ratio})

    df = run(species, q.parameters, **kwargs)
    df = df.assign(exchange = 'medium exchange, k_ex_f = 0.01')
    dfs.append(df)
    
    # run with fast exchange
    if not np.isfinite(NMR_ratio):  # if biased state 2
        q.parameters.update({'k_ex_f':200, 'k_ex_r':2})
    elif NMR_ratio == 0.0:  # if biased state 1
        q.parameters.update({'k_ex_f':2, 'k_ex_r':200})
    else: 
        q.parameters.update({'k_ex_f':200, 'k_ex_r':200/NMR_ratio})

    df = run(species, q.parameters, **kwargs)
    df = df.assign(exchange = 'fast exchange, k_ex_f = 200')
    dfs.append(df)
    
    df = pd.concat(objs=dfs).reset_index(drop=True)
    df = df.assign(mutant = mutant,
                   t = df.t/60,
                   Gsp1GDP = df.Gsp1GDP/total_Gsp1,
                   frac_II = df.Gsp1GTP_II/(df.Gsp1GTP_I+df.Gsp1GTP_II))
    return df


# import the rest of the mutant kinetic parameters
GAP_kinetics = pd.read_csv('GAP_kinetics.csv')

# import the NMR data
NMR = pd.read_csv('31P_NMR_Data.csv')

NMR = (NMR
      .loc[NMR.peak.isin(['gamma1', 'gamma2'])]
      [['variant', 'peak', 'integral_rel']]
      .pivot(index='variant', columns='peak', values='integral_rel')
      .assign(gamma1 = lambda df: df.gamma1.apply(lambda g1: 0 if pd.isnull(g1) else g1),
              gamma2 = lambda df: df.gamma2.apply(lambda g2: 0 if pd.isnull(g2) else g2))
      .assign(ratio = lambda df: df.gamma2/df.gamma1,
              frac_II = lambda df: df.gamma2/(df.gamma1+df.gamma2))
      .reset_index()
       [['variant','ratio','frac_II']]
      .rename(columns={'variant':'mutant'})
      )

mutant_params = pd.merge(GAP_kinetics, NMR)

# run all mutants
dfs = []
for row in mutant_params.itertuples(index=False):
    dfs.append(run_mutant(*row))
    

df = pd.concat(dfs).reset_index(drop=True)
```

```python
sns.set(style="whitegrid", context="talk")

# plot all mutants on the same plot
g = sns.relplot(data = df, x = 't', y = 'Gsp1GDP',
                col='exchange', hue='mutant', kind='line',
                hue_order = list(NMR.sort_values('ratio').mutant),
                linewidth=2, legend='full',
                height = 7, aspect=1)
g.set(xlabel='Time (min)', ylabel='Fraction Gsp1GTP hydrolyzed', xlim=(0,100))
plt.savefig('plots/mutants_progress.png')

g = sns.relplot(data = df, x = 't', y = 'frac_II',
                col='exchange', hue='mutant', kind='line',
                hue_order = list(NMR.sort_values('ratio').mutant),
                linewidth=2, legend='full',
                height = 7, aspect=1)
g.set(xlabel='Time (min)', ylabel='Fraction remaining Gsp1GTP in State II',
      xlim=(0,100), ylim=(0,1))
plt.savefig('plots/mutants_frac_II.png')

g = sns.relplot(data = df, x = 't', y = 'Gsp1GDP',
                col='mutant', hue='exchange', kind='line',
                col_wrap = 3, col_order = list(NMR.sort_values('ratio').mutant),
                linewidth=2, legend='full',
                height = 4, aspect=1)
g.set(xlabel='Time (min)', ylabel='\nFraction Gsp1GTP hydrolyzed', xlim=(0,100))
plt.tight_layout()
g._legend.set_bbox_to_anchor([0.975, 0.1])
plt.savefig('plots/mutants_sep.png')

g = sns.relplot(data = df, x = 't', y = 'frac_II',
                col='mutant', hue='exchange', kind='line',
                col_wrap = 3, col_order = list(NMR.sort_values('ratio').mutant),
                linewidth=2, legend='full',
                height = 4, aspect=1)
g.set(xlabel='Time (min)', ylabel='Fraction remaining\nGsp1GTP in State II',
      xlim=(0,100), ylim=(0,1))
plt.tight_layout()
g._legend.set_bbox_to_anchor([0.975, 0.1])
plt.savefig('plots/mutants_sep_frac_II.png')
```
